```python
combined_table = pd.concat([squinky_table.T,squinky_lexical_table.T,squinky_scores_table.T,squinky_reading_table.T,squinky_nlp_table.T,total_table.T])
combined_table = combined_table.abs()
combined_table
```

```python
def significance_table(r,err,significances):
    nothing  = np.array(combined_table.loc[[r + " Regression + Nothing: " + err]])[0]
    lexical = np.array(combined_table.loc[[r + " Regression + Lexical: " + err]])[0]
    scores = np.array(combined_table.loc[[r + " Regression + Scores: " + err]])[0]
    reading = np.array(combined_table.loc[[r +  " Regression + Reading: " + err]])[0]
    nlp = np.array(combined_table.loc[[r + " Regression + NLP: " + err]])[0]
    all = np.array(combined_table.loc[[r + " Regression + All: " + err]])[0]


    significances["Base - Lexical | " + r + " | " + err] = stats.ttest_ind(nothing, lexical)[0], stats.ttest_ind(nothing, lexical)[1] ,(stats.ttest_ind(nothing, lexical)[1] < 0.05)
    significances["Base - Scores | " + r + " | " + err] = stats.ttest_ind(nothing, scores)[0],stats.ttest_ind(nothing, scores)[1] ,(stats.ttest_ind(nothing, scores)[1] < 0.05)
    significances["Base - Reading | " + r + " | " + err] = stats.ttest_ind(nothing, reading)[0],stats.ttest_ind(nothing, reading)[1] ,(stats.ttest_ind(nothing, reading)[1] < 0.05)
    significances["Base - NLP | " + r + " | " + err] = stats.ttest_ind(nothing, nlp)[0],stats.ttest_ind(nothing, nlp)[1] ,(stats.ttest_ind(nothing, nlp)[1] < 0.05)
    significances["Base - All | " + r + " | " + err] = stats.ttest_ind(nothing, all)[0],stats.ttest_ind(nothing, all)[1] ,(stats.ttest_ind(nothing, all)[1] < 0.05)

    return significances
```

```python
significances = pd.DataFrame()
significances = significance_table("Linear","MSE",significances)
significances = significance_table("Linear","MAE",significances)
significances = significance_table("Ridge","MSE",significances)
significances = significance_table("Ridge","MAE",significances)
significances = significance_table("Lasso","MSE",significances)
significances = significance_table("Lasso","MAE",significances)
significances = significance_table("LARS Lasso","MSE",significances)
significances = significance_table("LARS Lasso","MAE",significances)
significances = significance_table("Bayesian Ridge","MSE",significances)
significances = significance_table("Bayesian Ridge","MAE",significances)
significances = significance_table("SGD","MSE",significances)
significances = significance_table("SGD","MAE",significances)
significances = significances.T
```
