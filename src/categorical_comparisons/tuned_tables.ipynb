```python
# Squinky + Scores 
squinky = pd.read_csv("..\..\data\mturk_experiment_2.csv",e ding="unicode escape")
scores = pd.read_csv("..\..\data\squinky_scores.csv",encoding="unicode escape")
squinky_scores = pd.merge(squinky,scores)

feature_names = ["Informativeness","Implicature","Length in Words","Length in Characters","F-score","I-score","Lexical Density","ADF Score","FK Grade Level","Reading Time","WF Score"]
squinky_scores_table = regression_table_tuned(squinky_scores,feature_names,"Scores")
squinky_scores_table
```

```python
# Squinky + Reading

squinky = pd.read_csv("..\..\data\mturk_experiment_2.csv",encoding="unicode escape")
reading = pd.read_csv("..\..\data\squinky_reading.csv",encoding="unicode escape")
squinky_reading = pd.merge(squinky,reading)

feature_names = ["Informativeness","Implicature","Length in Words","Length in Characters","F-score","I-score","Lexical Density","FK Reading Ease","FOG Scale","SMOG Index","ARI","CL Index","LW Formula","DC Score","Readability Consensus","Spache Formula"]
squinky_reading_table = regression_table_tuned(squinky_reading,feature_names,"Reading")
squinky_reading_table
```

```python
# Squinky + NLP

squinky = pd.read_csv("..\..\data\mturk_experiment_2.csv",encoding="unicode escape")
nlp = pd.read_csv("..\..\data\squinky_nlp.csv")
squinky_nlp = pd.merge(squinky,nlp)

feature_names = ["Informativeness","Implicature","Length in Words","Length in Characters","F-score","I-score","Lexical Density","Cluster","Topic"]
squinky_nlp_table = regression_table_tuned(squinky_nlp,feature_names,"NLP")
squinky_nlp_table
```

```python
data_frames = [squinky, lexical, scores, reading, nlp]
total = pd.concat(data_frames, axis=1)

feature_names = ['Informativeness', 'Implicature',
                 'Length in Words', 'Length in Characters', 'F-score', 'I-score',
                 'Lexical Density', 'Adjective Density',
                 'Spellcheck Percentage', 'Syllable Ratio', 'Latinate vs Germanic', 'ADF Score', 'FK Grade Level', 'Reading Time', 'WF Score', 'FK Reading Ease', 'FOG Scale', 'SMOG Index', 'ARI',
                 'CL Index', 'LW Formula', 'DC Score', 'Readability Consensus',
                 'Spache Formula','Cluster', 'Topic']


total_table = regression_table_tuned(total,feature_names,"All")
total_table
```
