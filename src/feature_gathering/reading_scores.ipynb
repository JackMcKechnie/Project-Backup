```python
import pandas as pd
import textstat
```

```python
# Import SQUINKY sentences
squinky = pd.read_csv("..\..\data\mturk_experiment_2.csv",encoding="unicode escape")
sentences = pd.DataFrame()
sentences["Sentence"] = squinky["Sentence"]
```

```python
sentences["FK Reading Ease"] = sentences["Sentence"].apply(textstat.flesch_reading_ease)
sentences["FOG Scale"] = sentences["Sentence"].apply(textstat.gunning_fog)
sentences["SMOG Index"] = sentences["Sentence"].apply(textstat.smog_index)
sentences["ARI"] = sentences["Sentence"].apply(textstat.automated_readability_index)
sentences["CL Index"] = sentences["Sentence"].apply(textstat.coleman_liau_index)
sentences["LW Formula"] = sentences["Sentence"].apply(textstat.linsear_write_formula)
sentences["DC Score"] = sentences["Sentence"].apply(textstat.dale_chall_readability_score)
sentences["Readability Consensus"] = sentences["Sentence"].apply(textstat.text_standard,float_output=True)
sentences["Spache Formula"] = sentences["Sentence"].apply(textstat.spache_readability)

#sentences.to_csv("..\..\data\squinky_reading.csv",index=False)
sentences
```
