```python
# Imports 
import pandas as pd
from f_score import f_score
from adjective_density import adjective_density
import textstat
from wf_score import wf_score
```

```python
# Import SQUINKY sentences
squinky = pd.read_csv("..\..\data\mturk_experiment_2.csv",encoding="unicode escape")
sentences = pd.DataFrame()
sentences["Sentence"] = squinky["Sentence"]
```

```python
sentences["ADF Score"] = sentences["Sentence"].apply(adjective_density)
sentences["FK Grade Level"] = sentences["Sentence"].apply(textstat.flesch_kincaid_grade)
sentences["Reading Time"] = sentences["Sentence"].apply(textstat.reading_time,ms_per_char=14.69)
sentences["WF Score"] = sentences["Sentence"].apply(wf_score)


#sentences.to_csv("..\..\data\squinky_scores.csv",index=False)
sentences
```
