```python
import pandas as pd
import tensorflow as tf
import numpy as np
from sklearn.model_selection import train_test_split
import gensim
```

```python
data = pd.read_csv("../../data/sensitivity_data/sensitivity_dataset.csv")
data = data[["Document","Sensitivity"]]

train_x, test_x, train_y, test_y = train_test_split(data['Document'],data['Sensitivity'],test_size=0.00001,random_state=5)
train_x = np.array(train_x)
train_y = np.array(train_y)
test_x = np.array(test_x)
test_y = np.array(test_y)
```

```python
def preprocess(corpus,testing):
    for i, doc in enumerate(corpus):
        tokens = gensim.utils.simple_preprocess(doc)
        if testing == True:
            yield tokens
        else:
            yield gensim.models.doc2vec.TaggedDocument(tokens, [train_y[i]])
```

```python
train_corpus = list(preprocess(train_x,testing=False))
```

```python
doc2vec_model = gensim.models.doc2vec.Doc2Vec(vector_size=5000, min_count=2, epochs=40)
doc2vec_model.build_vocab(train_corpus)
doc2vec_model.train(train_corpus, total_examples=doc2vec_model.corpus_count, epochs=doc2vec_model.epochs)
```

```python
doc_vectors = []
for i,doc in enumerate(train_x):
    vect = list(doc2vec_model.infer_vector(gensim.utils.simple_preprocess(doc)))
    doc_vectors.append([vect,train_y[i]])
    print(str(i + 1) + " Documents Processed..." + str(round( (i/len(train_x)) * 100, 2)) + "%")
```

```python
df = pd.DataFrame(doc_vectors)
df.columns = ["Vector","Sensitivity"]
df.to_csv("doc2vec_embeddings.csv")
```
