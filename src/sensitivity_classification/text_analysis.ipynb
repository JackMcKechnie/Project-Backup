```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("../../data/sensitivity_data/sensitivity_dataset.csv")
data = data[["Document","Sensitivity"]]
```

```python
word_length = []
for doc in data["Document"]:
    word_length.append(len(doc.split(" ")))
```

```python
plt.hist(word_length,bins=int(max(word_length)/20))
plt.xlabel("Document Length")
plt.ylabel("Frequency")
plt.title("Length of Documents")
```

```python
plt.boxplot(word_length,vert=False)
plt.xlabel("Document Length")
plt.ylabel("Frequency")
plt.title("Length of Documents")
plt.yticks([1],[""])
```

```python
small_docs = 0
large_docs = 0
for doc in word_length:
    if doc <= 1024:
        small_docs += 1
    else:
        large_docs += 1

plt.bar(["<= 1024","> 1024"],[small_docs,large_docs])
plt.xlabel("Document Length")
plt.ylabel("Frequency")
plt.title("Length of Documents")
```
