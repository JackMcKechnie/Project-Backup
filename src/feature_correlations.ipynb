```python
#Imports 
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
import dataframe_image as dfi
```

```python
data = pd.read_csv("data\extended_mturk.csv")
data = data[["Formality","Informativeness","Implicature","Length in Words","Length in Characters","F-score","I-score","Lexical Density","Readability","Flesch-Kincaid","ADF Score","S+G Percent"]]
```

```python
#Get correlation matrix
corr_matrix = data.corr()
```

```python
#Plot correlation matrix as a heatmap
plt.subplots(figsize=(20,15))
plt.title("Heatmap of Correlation Matrix Of Proposed Features \n",size=28)
sn.heatmap(corr_matrix, annot=True)
plt.savefig('foo.png', bbox_inches='tight')
```

```python
#Save correlation matrix
df_styled = corr_matrix.style.background_gradient()
dfi.export(df_styled,"data/correlation_matrix.png")
```
